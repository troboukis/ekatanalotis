```python
df['cat_clean'] = df.category_name.apply(lambda x: re.split(",", x)[-1])
```

```python
price_per_year_category = df.groupby(["cat_clean", df.date.dt.year]).price.median().unstack().reset_index()
```

```python
# Διαφορά ανάμεσα σε mean & median
comparison = df.groupby([df['date'].dt.year, 'cat_clean'])['price'].agg(['mean', 'median'])
comparison['diff_pct'] = round((comparison['mean'] - comparison['median']) / comparison['median'] * 100, 1)
comparison.sort_values('diff_pct', ascending=False)
```

```python
price_per_year_category['pct_change_25_26'] = round((price_per_year_category[2026] - price_per_year_category[2025])/price_per_year_category[2025]*100, 1)

price_per_year_category['pct_change_24_26'] = round((price_per_year_category[2026] - price_per_year_category[2024])/price_per_year_category[2024]*100, 1)
```

```python
import numpy as np
print("direction_25_26\n")
price_per_year_category['direction_25_26'] = np.sign(price_per_year_category['pct_change_25_26']).map({1.0: 'Αύξηση', -1.0: 'Μείωση', 0.0: 'Αμετάβλητη'})
print(price_per_year_category['direction_25_26'].value_counts())

print("direction_24_26\n")
price_per_year_category['direction_24_26'] = np.sign(price_per_year_category['pct_change_24_26']).map({1.0: 'Αύξηση', -1.0: 'Μείωση', 0.0: 'Αμετάβλητη'})
price_per_year_category['direction_24_26'].value_counts()
```

```python
from scipy.stats import linregress
import numpy as np

# Μέση μηνιαία τιμή ανά κατηγορία
monthly = (
    df.groupby([df['date'].dt.to_period('M'), 'cat_clean'])['price']
    .mean()
    .reset_index()
)
monthly['month_num'] = monthly['date'].apply(lambda x: x.ordinal)

# Slope ανά κατηγορία
def calc_slope(g):
    slope, intercept, r, p, se = linregress(g['month_num'], g['price'])
    return pd.Series({'slope_per_month': round(slope, 4), 'r_squared': round(r**2, 3)})

growth_rate = monthly.groupby('cat_clean').apply(calc_slope).reset_index()
growth_rate.sort_values('slope_per_month', ascending=False)
```

```python
''' 
Ερμηνεία:
slope = 0.05 → η μέση τιμή της κατηγορίας αυξάνεται κατά 0.05€ τον μήνα
slope = -0.03 → μειώνεται κατά 0.03€ τον μήνα
slope ≈ 0 → σταθερή τιμή
Το r_squared (R²) δείχνει πόσο καλά η ευθεία περιγράφει τα δεδομένα:

R² ≈ 1.0 → η αύξηση είναι σταθερή και γραμμική (η τιμή ανεβαίνει σταδιακά κάθε μήνα)
R² ≈ 0.5 → υπάρχει τάση αλλά με αρκετή διακύμανση
R² ≈ 0.0 → δεν υπάρχει γραμμική τάση — η τιμή πηδάει πάνω-κάτω, οπότε το slope δεν είναι αξιόπιστο
'''
# Κατηγορίες με σταθερή αύξηση (slope > 0 και R² > 0.5)
growth_rate[(growth_rate['slope_per_month'] > 0.1) & (growth_rate['r_squared'] > 0.5)] \
    .sort_values('slope_per_month', ascending=False)
```

```python
# CAGR — Compound Annual Growth Rate
'''
CAGR — Compound Annual Growth Rate
Ο CAGR (Σύνθετος Ετήσιος Ρυθμός Αύξησης) είναι ο σταθερός ετήσιος ρυθμός που θα χρειαζόταν για να πας από μια αρχική τιμή σε μια τελική, σε ένα δεδομένο χρονικό διάστημα.

Τύπος

CAGR = (τελική_τιμή / αρχική_τιμή) ^ (1 / αριθμός_ετών) - 1
Παράδειγμα
Μια κατηγορία κοστίζει 10€ το 2024 και 12.10€ το 2026 (2 χρόνια):


CAGR = (12.10 / 10) ^ (1/2) - 1 = 1.21 ^ 0.5 - 1 = 1.1 - 1 = 0.10 → 10%
Δηλαδή: αν η τιμή αυξανόταν σταθερά 10% κάθε χρόνο, θα πήγαινε:

2024: 10.00€
2025: 11.00€ (+10%)
2026: 12.10€ (+10%)
CAGR vs Slope vs Απλή ποσοστιαία μεταβολή
Τι μετράει	Μονάδα	Πλεονέκτημα	Αδυναμία
% μεταβολή	Διαφορά αρχής→τέλους	%	Απλό	Αγνοεί τι γίνεται ενδιάμεσα
CAGR	Εξομαλυσμένος ετήσιος ρυθμός	% ετησίως	Συγκρίσιμο μεταξύ κατηγοριών με διαφορετικά χρονικά διαστήματα	Αγνοεί κι αυτό τη διαδρομή (π.χ. αν ανέβηκε πολύ και μετά έπεσε)
Slope	Ευρώ αύξηση ανά μήνα	€/μήνα	Δείχνει αν η αύξηση είναι σταθερή (μέσω R²)	Δεν είναι % — δύσκολο να συγκρίνεις ακριβά vs φθηνά προϊόντα
Πότε να χρησιμοποιήσεις τι
CAGR: Θέλεις να συγκρίνεις ρυθμούς αύξησης μεταξύ κατηγοριών ως ποσοστό — π.χ. "ποια κατηγορία αυξάνεται πιο γρήγορα ετησίως;"
Slope + R²: Θέλεις να δεις αν η αύξηση είναι πραγματική τάση ή τυχαία διακύμανση
Και τα δύο μαζί: Ιδανικό — CAGR για το "πόσο" και R² για το "πόσο σίγουροι είμαστε"
Προσοχή
Το CAGR βλέπει μόνο αρχή και τέλος. Αν μια κατηγορία πήγε 10€→20€→12€, το CAGR θα δείξει μικρή αύξηση, αλλά στην πραγματικότητα είχε τεράστια αστάθεια. Γι' αυτό ο συνδυασμός με R² είναι πιο αξιόπιστος.
'''
```

```python
n_years = 2  # 2024 → 2026

cagr = (
    (price_per_year_category[2026] / price_per_year_category[2024]) ** (1 / n_years) - 1
) * 100

price_per_year_category['cagr'] = round(cagr, 2)
price_per_year_category.sort_values('cagr', ascending=False)
```

```python
price_analysis = pd.merge(price_per_year_category, growth_rate, how="inner", on="cat_clean")
```
