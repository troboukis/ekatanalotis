```python
# ============================================
# ΑΝΑΛΥΣΗ 2: Price Index - Εξέλιξη τιμών (Βάση: Οκτ 2024 = 100)
# ============================================

import matplotlib.pyplot as plt

# Βρες τον πρώτο μήνα (Οκτ 2024)
base_period = df['date'].dt.to_period('M').min()

# Υπολόγισε τη ΔΙΑΜΕΣΗ τιμή κάθε προϊόντος στη βάση (Οκτ 2024)
base_prices = df[df['date'].dt.to_period('M') == base_period].groupby('product_id')['price'].median()

# Merge με το df για να έχουμε τη base price για κάθε προϊόν
df_indexed = df.merge(base_prices.rename('base_price'), on='product_id', how='inner')

# Υπολόγισε το price index για κάθε εγγραφή
df_indexed['price_index'] = (df_indexed['price'] / df_indexed['base_price']) * 100

# OUTLIER DETECTION: Percentile method (1%-99%)
lower_bound = df_indexed['price_index'].quantile(0.01)
upper_bound = df_indexed['price_index'].quantile(0.99)

df_clean = df_indexed[(df_indexed['price_index'] >= lower_bound) & 
                       (df_indexed['price_index'] <= upper_bound)]

outlier_pct = (len(df_indexed) - len(df_clean)) / len(df_indexed) * 100
print(f"Αφαιρέθηκαν {len(df_indexed) - len(df_clean):,} outliers ({outlier_pct:.1f}%)")
print(f"Εύρος τιμών που κρατήθηκαν: {lower_bound:.0f} - {upper_bound:.0f}")

# ΔΙΑΜΕΣΟΣ price index ανά μήνα
monthly_index = df_clean.groupby(df_clean['date'].dt.to_period('M'))['price_index'].median()

# Plot
fig, ax = plt.subplots(figsize=(14, 6))

# Μετατροπή Period σε string για σωστή απεικόνιση
months = [str(m) for m in monthly_index.index]
values = monthly_index.values

ax.plot(months, values, marker='o', linewidth=2, markersize=8, color='#2E86AB', label='Price Index')
ax.axhline(y=100, color='gray', linestyle='--', alpha=0.7, label='Βάση (Οκτ 2024)')
ax.fill_between(months, 100, values, alpha=0.3, color='#2E86AB')

ax.set_title('Δείκτης Τιμών (Median) - Βάση: Οκτώβριος 2024 = 100', fontsize=14, fontweight='bold')
ax.set_xlabel('Μήνας', fontsize=12)
ax.set_ylabel('Price Index', fontsize=12)
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left')

# Rotate x labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Στατιστικά
print(f"\n=== Στατιστικά Price Index (Median) ===")
print(f"Βάση: {base_period} = 100")
print(f"Τελευταίος μήνας: {monthly_index.index[-1]} = {monthly_index.iloc[-1]:.2f}")
print(f"Συνολική μεταβολή: {monthly_index.iloc[-1] - 100:+.2f}%")
print(f"\nMin: {monthly_index.min():.2f} ({monthly_index.idxmin()})")
print(f"Max: {monthly_index.max():.2f} ({monthly_index.idxmax()})")
```

```python
# ============================================
# ΑΝΑΛΥΣΗ 4: Ανάλυση τιμών ανά Supermarket (Interactive)
# ============================================

import ipywidgets as widgets
from IPython.display import display, clear_output

# Δημιουργία main_category αν δεν υπάρχει
if 'main_category' not in df.columns:
    df['main_category'] = df['category_name'].str.split(',').str[0].str.strip()

# Λίστα supermarkets
supermarkets = sorted(df['merchant'].unique())

def analyze_supermarket(merchant):
    clear_output(wait=True)
    
    # Display dropdown again
    display(dropdown)
    
    # Filter data για το επιλεγμένο supermarket
    df_merchant = df[df['merchant'] == merchant].copy()
    
    print(f"\n{'='*60}")
    print(f"ΑΝΑΛΥΣΗ: {merchant.upper()}")
    print(f"{'='*60}")
    print(f"Συνολικές εγγραφές: {len(df_merchant):,}")
    print(f"Μοναδικά προϊόντα: {df_merchant['product_id'].nunique():,}")
    print(f"Περίοδος: {df_merchant['date'].min().strftime('%d/%m/%Y')} - {df_merchant['date'].max().strftime('%d/%m/%Y')}")
    
    # --- Γραφήματα ---
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Top Left: Price Index για αυτό το supermarket
    base_period = df_merchant['date'].dt.to_period('M').min()
    base_prices = df_merchant[df_merchant['date'].dt.to_period('M') == base_period].groupby('product_id')['price'].median()
    df_m_indexed = df_merchant.merge(base_prices.rename('base_price'), on='product_id', how='inner')
    df_m_indexed['price_index'] = (df_m_indexed['price'] / df_m_indexed['base_price']) * 100
    
    # Remove outliers (1%-99%)
    lb = df_m_indexed['price_index'].quantile(0.01)
    ub = df_m_indexed['price_index'].quantile(0.99)
    df_m_clean = df_m_indexed[(df_m_indexed['price_index'] >= lb) & (df_m_indexed['price_index'] <= ub)]
    
    monthly_idx = df_m_clean.groupby(df_m_clean['date'].dt.to_period('M'))['price_index'].median()
    months = [str(m) for m in monthly_idx.index]
    axes[0,0].plot(months, monthly_idx.values, marker='o', linewidth=2, color='#2E86AB')
    axes[0,0].axhline(y=100, color='gray', linestyle='--', alpha=0.7)
    axes[0,0].fill_between(months, 100, monthly_idx.values, alpha=0.3)
    axes[0,0].set_title(f'Price Index - {merchant}')
    axes[0,0].set_ylabel('Index (Βάση=100)')
    axes[0,0].tick_params(axis='x', rotation=45)
    # Εμφάνιση κάθε 2ου label
    for i, label in enumerate(axes[0,0].xaxis.get_ticklabels()):
        if i % 2 != 0:
            label.set_visible(False)
    
    # Top Right: Κατανομή τιμών
    axes[0,1].hist(df_merchant['price'].clip(upper=df_merchant['price'].quantile(0.95)), 
                   bins=50, edgecolor='black', alpha=0.7, color='#28A745')
    axes[0,1].axvline(df_merchant['price'].median(), color='red', linestyle='--', 
                      label=f"Διάμεση: €{df_merchant['price'].median():.2f}")
    axes[0,1].set_title(f'Κατανομή τιμών - {merchant}')
    axes[0,1].set_xlabel('Τιμή (€)')
    axes[0,1].set_ylabel('Συχνότητα')
    axes[0,1].legend()
    
    # Bottom Left: Top κατηγορίες (by median price)
    cat_prices = df_merchant.groupby('main_category')['price'].median().sort_values(ascending=False).head(10)
    axes[1,0].barh(range(len(cat_prices)), cat_prices.values, color='#FF6B6B')
    axes[1,0].set_yticks(range(len(cat_prices)))
    axes[1,0].set_yticklabels([c[:20] + '...' if len(c) > 20 else c for c in cat_prices.index])
    axes[1,0].set_title(f'Top 10 ακριβότερες κατηγορίες - {merchant}')
    axes[1,0].set_xlabel('Διάμεση τιμή (€)')
    
    # Bottom Right: Προσφορές & Μόνιμες μειώσεις
    promo_pct = df_merchant['promo'].mean() * 100
    monimi_pct = df_merchant['monimi_meiosi'].mean() * 100
    regular_pct = 100 - promo_pct - monimi_pct
    
    labels = ['Κανονική τιμή', 'Προσφορά', 'Μόνιμη μείωση']
    sizes = [regular_pct, promo_pct, monimi_pct]
    colors = ['#4ECDC4', '#FFE66D', '#FF6B6B']
    axes[1,1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
    axes[1,1].set_title(f'Τύποι τιμών - {merchant}')
    
    plt.tight_layout()
    plt.show()
    
    # Στατιστικά σύγκρισης με τους υπόλοιπους
    print(f"\n--- Σύγκριση με υπόλοιπα supermarkets ---")
    all_median = df.groupby('merchant')['price'].median().sort_values()
    rank = list(all_median.index).index(merchant) + 1
    print(f"Κατάταξη (φθηνότερο→ακριβότερο): {rank}/{len(all_median)}")
    print(f"Διάμεση τιμή {merchant}: €{all_median[merchant]:.2f}")
    print(f"Διάμεση τιμή αγοράς: €{df['price'].median():.2f}")

# Create dropdown
dropdown = widgets.Dropdown(
    options=supermarkets,
    value=supermarkets[0],
    description='Supermarket:',
    style={'description_width': 'initial'}
)

# Connect dropdown to function
dropdown.observe(lambda change: analyze_supermarket(change['new']), names='value')

# Display and run initial analysis
display(dropdown)
analyze_supermarket(supermarkets[0])
```

```python
# ============================================
# ΑΝΑΛΥΣΗ 5: Κατηγορίες με μεγαλύτερη αύξηση τιμών
# ============================================
# Χρησιμοποιούμε MEDIAN για ανθεκτικότητα σε outliers

# Εξαγωγή πρώτης κατηγορίας (top-level category)
df['main_category'] = df['category_name'].str.split(',').str[0].str.strip()

# Μέση τιμή ανά κατηγορία στην αρχή και στο τέλος της περιόδου
first_month = df['date'].dt.to_period('M').min()
last_month = df['date'].dt.to_period('M').max()

# ΔΙΑΜΕΣΗ τιμή ανά κατηγορία
prices_first = df[df['date'].dt.to_period('M') == first_month].groupby('main_category')['price'].median()
prices_last = df[df['date'].dt.to_period('M') == last_month].groupby('main_category')['price'].median()

# Υπολογισμός ποσοστιαίας μεταβολής
price_change = pd.DataFrame({
    'price_start': prices_first,
    'price_end': prices_last
}).dropna()
price_change['change_pct'] = (price_change['price_end'] / price_change['price_start'] - 1) * 100

# Ταξινόμηση
price_change_sorted = price_change.sort_values('change_pct', ascending=False)

print(f"=== Μεταβολή τιμών ανά κατηγορία - MEDIAN ({first_month} → {last_month}) ===\n")
print("Top 10 κατηγορίες με ΑΥΞΗΣΗ:")
print(price_change_sorted.head(10)[['price_start', 'price_end', 'change_pct']].round(2))

print("\nTop 10 κατηγορίες με ΜΕΙΩΣΗ:")
print(price_change_sorted.tail(10)[['price_start', 'price_end', 'change_pct']].round(2))

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Top increases
top_inc = price_change_sorted.head(10)
colors_inc = ['#d62728' if x > 0 else '#2ca02c' for x in top_inc['change_pct']]
axes[0].barh(range(len(top_inc)), top_inc['change_pct'], color=colors_inc)
axes[0].set_yticks(range(len(top_inc)))
axes[0].set_yticklabels([c[:25] + '...' if len(c) > 25 else c for c in top_inc.index])
axes[0].set_xlabel('Μεταβολή τιμής (%)')
axes[0].set_title('Top 10 κατηγορίες με ΑΥΞΗΣΗ')
axes[0].axvline(0, color='black', linewidth=0.5)

# Right: Top decreases
top_dec = price_change_sorted.tail(10).iloc[::-1]
colors_dec = ['#d62728' if x > 0 else '#2ca02c' for x in top_dec['change_pct']]
axes[1].barh(range(len(top_dec)), top_dec['change_pct'], color=colors_dec)
axes[1].set_yticks(range(len(top_dec)))
axes[1].set_yticklabels([c[:25] + '...' if len(c) > 25 else c for c in top_dec.index])
axes[1].set_xlabel('Μεταβολή τιμής (%)')
axes[1].set_title('Top 10 κατηγορίες με ΜΕΙΩΣΗ')
axes[1].axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```
